```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("credit_risk.csv")
df.head()
```

```python
pd.crosstab(df['residence_status'], df['status'], normalize='index') * 100
```

```python
(pd.crosstab(
    df['residence_status'],
    df['status'],
    normalize='index'
)[1] * 100).plot(kind='bar')

import matplotlib.pyplot as plt
plt.title('Default Rate by Residence Status')
plt.ylabel('Default Rate (%)')
plt.xlabel('Residence Status')
plt.show()
```

```python
pd.crosstab(df['employment'], df['status'], normalize='index') * 100
```

```python
(pd.crosstab(
    df['employment'],
    df['status'],
    normalize='index'
)[1] * 100).plot(kind='bar')

import matplotlib.pyplot as plt
plt.title('Default Rate by Employment Status')
plt.ylabel('Default Rate (%)')
plt.xlabel('Employment Status')
plt.show()
```
