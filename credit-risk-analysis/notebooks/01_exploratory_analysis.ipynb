```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("credit_risk.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df['status'].value_counts()
df['status'].value_counts(normalize=True) * 100
```

```python
import matplotlib.pyplot as plt
df['status'].value_counts().plot(kind='bar')
plt.title('Distribution of Credit Status')
plt.xlabel('Status (0 = No Default, 1 = Default)')
plt.ylabel('Number of Customers')
plt.show()
```

```python
df.groupby('status')['customer_age'].mean()
```

```python
df.groupby('status')['income'].mean()
```

```python
df.groupby('status')['months_with_bank'].mean()
```

```python
df.boxplot(column='income', by='status')
plt.title('Income Distribution by Credit Status')
plt.suptitle('')
plt.xlabel('Status')
plt.ylabel('Income')
plt.show()
```
