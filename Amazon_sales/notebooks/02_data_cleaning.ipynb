# Amazon Sales — Data Cleaning

Neste notebook realizamos a limpeza, padronização e tratamento dos dados.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../DB/amazon_sales_data_2025.csv')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df = df.rename(columns={'date': 'order_date'})
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['calculated_sales'] = df['price'] * df['quantity']

(df['total_sales'] != df['calculated_sales']).sum()
```

```python
df = df.rename(columns={'total_sales': 'revenue'})
df.drop(columns='calculated_sales', inplace=True)
```

```python
df['price'] = df['price'].astype(float)
df['quantity'] = df['quantity'].astype(int)
df['revenue'] = df['revenue'].astype(float)
```

```python
df.to_csv('../DB/amazon_sales_data_2025.csv', index=False)
```
